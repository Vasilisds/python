```python
df = df[df['Order Date'].str[:2]!='Or']
```

```python
df['Month']=df['Month'].astype("int32")
```

```python
df.head()
```

```python
df['Quantity Ordered'] = df['Quantity Ordered'].astype("int32")
df['Price Each'] = df['Price Each'].astype("float")
```

```python
## Add sales column
df['Sales']= df.loc[:,'Quantity Ordered'] * df.loc[:,'Price Each']
```

```python
df.dtypes
```

```python
##Best Selling Month 

results = df.groupby("Month")['Sales'].sum()
```

```python
import matplotlib.pyplot as plt
months = range(1,13)
plt.bar(months,results)
plt.show()
```

```python
## What City had the highest number of sales
def get_city(x):
    return x.split(',')[1]

def get_code(x):
    return x.split(',')[2][:3]
```

```python
df['City'] = df['Purchase Address'].apply(lambda x : get_city(x) + " "+ get_code(x))
```
